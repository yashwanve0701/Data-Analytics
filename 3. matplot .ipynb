```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,100)# that means we will get 100 values with space of 10 bwtween 0 to 10
y=np.sin(x)
z=np.cos(x)
```

```python
print(x)
```

```python
print(z)
```

```python
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel("sin value")
plt.title("plot")
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```
